```python
df[[costo_unitario, categoria]].groupby(categoria).describe()
medidas_catg = {}
for categoria_unica in categorias_unicas:
    moda = df[df[categoria] == categoria_unica][costo_unitario].mode()[0]
    mediana = df[df[categoria] == categoria_unica][costo_unitario].median()
    promedio = df[df[categoria] == categoria_unica][costo_unitario].mean()
    std = df[df[categoria] == categoria_unica][costo_unitario].std()
    medidas_catg[categoria_unica] = {'Moda': moda, 'Mediana': mediana, 'Promedio': promedio}

print(medidas_catg)
```

```python
# Creamos una copia del DataFrame para trabajar con los valores atípicos de costo_unitario


# Definimos umbrales fijos para detección de valores anormales
LIMITE_SUPERIOR = 10000
LIMITE_INFERIOR = 30

# Identificamos y reemplazamos valores atípicos por categoría
valores_reemplazados_altos = 0
valores_reemplazados_bajos = 0

for categoria_unica in categorias_unicas:
    # Identificamos valores que superan el límite superior
    mask_outliers_altos = (df[categoria] == categoria_unica) & (df[costo_unitario] > LIMITE_SUPERIOR)
    num_outliers_altos = mask_outliers_altos.sum()
    
    # Identificamos valores que están por debajo del límite inferior
    mask_outliers_bajos = (df[categoria] == categoria_unica) & (df[costo_unitario] < LIMITE_INFERIOR)
    num_outliers_bajos = mask_outliers_bajos.sum()
    
    # Obtenemos la moda de la categoría
    moda_categoria = medidas_catg[categoria_unica]['Moda']
    
    if num_outliers_altos > 0:
        # Reemplazamos valores altos con la moda
        df.loc[mask_outliers_altos, costo_unitario] = moda_categoria
        valores_reemplazados_altos += num_outliers_altos
        print(f"{categoria_unica}: {num_outliers_altos} valores > ${LIMITE_SUPERIOR} reemplazados con moda = ${moda_categoria:.2f}")
    
    if num_outliers_bajos > 0:
        # Reemplazamos valores bajos con la moda
        df.loc[mask_outliers_bajos, costo_unitario] = moda_categoria
        valores_reemplazados_bajos += num_outliers_bajos
        print(f"{categoria_unica}: {num_outliers_bajos} valores < ${LIMITE_INFERIOR} reemplazados con moda = ${moda_categoria:.2f}")

print(f"\n{'='*50}")
print(f"Total de valores altos (> ${LIMITE_SUPERIOR}) reemplazados: {valores_reemplazados_altos}")
print(f"Total de valores bajos (< ${LIMITE_INFERIOR}) reemplazados: {valores_reemplazados_bajos}")
print(f"Total general: {valores_reemplazados_altos + valores_reemplazados_bajos}")
print(f"\nEstadísticas después de la limpieza:")
print(df[costo_unitario].describe())
```

```python
# Reemplazamos valores "???" en la columna categoria basados en la mediana más cercana
df = df.copy()

# Calculamos la mediana de costo_unitario para cada categoría válida (excluyendo "???")
categorias_validas = [cat for cat in categorias_unicas if cat != '???']
medianas_por_categoria = {}

for cat in categorias_validas:
    mediana_cat = df[df[categoria] == cat][costo_unitario].median()
    medianas_por_categoria[cat] = mediana_cat

print("Medianas de costo_unitario por categoría:")
for cat, med in medianas_por_categoria.items():
    print(f"  {cat}: ${med:.2f}")

# Identificamos filas con categoria "???"
mask_desconocidos = df[categoria] == '???'
num_desconocidos = mask_desconocidos.sum()

print(f"\nTotal de registros con categoria '???': {num_desconocidos}")

# Reemplazamos cada "???" con la categoría cuya mediana esté más cercana
reemplazos_realizados = {}

for idx in df[mask_desconocidos].index:
    costo_valor = df.loc[idx, costo_unitario]
    
    # Encontramos la categoría con la mediana más cercana
    categoria_asignada = min(medianas_por_categoria.keys(), 
                            key=lambda cat: abs(medianas_por_categoria[cat] - costo_valor))
    
    # Reemplazamos la categoría
    df.loc[idx, categoria] = categoria_asignada
    
    # Contabilizamos los reemplazos
    reemplazos_realizados[categoria_asignada] = reemplazos_realizados.get(categoria_asignada, 0) + 1

print("\nCategorías asignadas a los registros '???' :")
for cat, count in reemplazos_realizados.items():
    print(f"  {cat}: {count} registros")

print(f"\nDistribución de categorías después del reemplazo:")
print(df[categoria].value_counts())
```
